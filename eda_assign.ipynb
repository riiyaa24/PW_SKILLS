```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('flight_price.xlsx.csv')
df.head(5)
```

```python
#dimensions of dataframe
rows, columns = df.shape
print("number of rows", rows)
print("number of columns", columns)
```

```python
#extract flight prices from the dataset
flight_price = df['Price']

#create a histogram to visualize distribution of flight price
plt.figure(figsize=(10,6))
plt.hist(flight_price, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#find the minimum and maximum prices
min_price = df['Price'].min()
max_price = df['Price'].max()

#calculate range of prices
price_range = max_price - min_price

#print results 
print(min_price)
print(max_price)
print(price_range)
```

```python
#create a boxplot of flight prices by airline using seaborn 
plt.figure(figsize=(12, 6))
sns.boxplot(x='Airline', y='Price', data=df, palette='Set2')
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

How outliers may impact analysis 
Potential impacts of outliers on analysis:
Outliers can skew statistical measures such as the mean and standard deviation, making them less representative of the central tendency and spread of the data.
Outliers can influence the results of data analysis, leading to biased estimates or incorrect conclusions.
Removing or addressing outliers may be necessary depending on the analysis and the context of the data.
Once potential outliers are identified, further analysis can be conducted to determine their impact on the dataset and decide whether they should be addressed or accounted for in subsequent analysis.
