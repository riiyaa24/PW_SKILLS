Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot: Scatter plots are used to visualize the relationship between two continuous variables. They are helpful in identifying patterns, clusters, or correlations between the variables.

Line Plot: Line plots are ideal for displaying data points connected by straight lines. They are often used to show trends or changes over time, making them suitable for time-series data analysis.

Histogram: Histograms are used to represent the distribution of a single continuous variable. They display the frequency of data points falling within specific bins or ranges, making it easy to identify the central tendency and spread of the data.

Bar Plot: Bar plots are useful for comparing categorical variables. They represent the frequency or proportion of data points within each category using vertical bars. Bar plots are effective in visualizing relationships between different categories or groups.

Heatmap: Heatmaps are graphical representations of data where values are depicted using colors. They are commonly used to visualize correlations between multiple variables in a tabular format. Heatmaps are especially useful for exploring relationships in large datasets and identifying patterns or clusters.

Seaborn provides convenient functions to create each of these plots with customizable features to enhance visualization and analysis.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot for different events and regions
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False)
plt.title("FMRI Signal over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots for age and fare with respect to passenger class (pclass)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram for price column with hue parameter for cut column
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack', bins=30, kde=True)
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title='Cut')
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris_data = sns.load_dataset("iris")

# Plot pair plot with hue parameter for species column
sns.pairplot(data=iris_data, hue='species', markers=['o', 's', 'D'])
plt.suptitle('Pair Plot of Iris Dataset by Species', y=1.02)
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Passenger Numbers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
