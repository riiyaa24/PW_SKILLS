Q1. What is difference between a t-test and a z-test? Provide an example scenario where you would use each type of test

Both t-tests and z-tests are statistical hypothesis tests used to determine if there is a significant difference between the means of two groups or if a sample mean differs significantly from a known population mean. However, they differ in the situations where they are applicable:

1. **T-test**:
   - The t-test is used when the sample size is small (typically less than 30) or when the population standard deviation is unknown.
   - It is also used when the population follows a normal distribution or when the sample size is sufficiently large (Central Limit Theorem applies).
   - Example scenario: Suppose you want to compare the mean scores of two groups of students (e.g., Group A and Group B) on a math test. If each group has fewer than 30 students or the population standard deviation of math scores is unknown, you would use a t-test.

2. **Z-test**:
   - The z-test is used when the sample size is large (typically greater than 30) and the population standard deviation is known.
   - It assumes that the population follows a normal distribution.
   - Example scenario: Imagine you want to test if a new drug has an effect on blood pressure. You collect blood pressure measurements from a large sample of individuals (e.g., 100 participants). If you know the population standard deviation of blood pressure, you can use a z-test to determine if there's a significant difference in blood pressure between those who took the drug and those who didn't.

Q2: Differentiate between one-tailed and two-tailed tests.

One-tailed and two-tailed tests are two types of hypothesis tests used in statistical analysis. They differ in how they assess the significance of the relationship between variables or the difference between groups. Here's a differentiation between them:

1. **One-tailed test**:
   - In a one-tailed test, the hypothesis is directional, meaning it specifies the direction of the effect or difference between groups.
   - It is used when the researcher has a specific expectation about the direction of the relationship or difference.
   - The critical region for the test is located entirely in one tail of the probability distribution.
   - Example: Suppose a researcher wants to test if a new teaching method improves test scores compared to the old method. The null hypothesis (H0) would be that there is no difference or that the new method is worse. The alternative hypothesis (H1) would be that the new method is better. This is a one-tailed test because it only considers the possibility of improvement with the new method.

2. **Two-tailed test**:
   - In a two-tailed test, the hypothesis is non-directional, meaning it does not specify the direction of the effect or difference between groups.
   - It is used when the researcher is interested in whether there is a significant difference between groups but does not have a specific expectation about the direction of the difference.
   - The critical region for the test is divided between both tails of the probability distribution.
   - Example: Consider a study examining whether there is a difference in reaction times between two groups of participants. The null hypothesis (H0) would be that there is no difference in reaction times between the groups. The alternative hypothesis (H1) would be that there is a difference, without specifying whether one group is faster or slower. This is a two-tailed test because it considers the possibility of a difference in either direction.

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

In hypothesis testing, Type I and Type II errors represent two different ways in which the conclusions drawn from the test can be incorrect:

1. **Type I Error**:
   - Type I error occurs when the null hypothesis (H0) is incorrectly rejected when it is actually true. In other words, it is a false positive.
   - The significance level (alpha, usually denoted as α) of the test determines the probability of making a Type I error. A lower significance level reduces the likelihood of a Type I error but increases the chance of a Type II error.
   - Example scenario: Suppose a medical researcher is testing a new drug to determine if it's effective in treating a certain disease. The null hypothesis (H0) is that the drug has no effect, i.e., it is no better than a placebo. If the researcher observes a statistically significant difference in the treatment group compared to the control group and rejects the null hypothesis, but in reality, the drug does not have any effect (H0 is true), this would be a Type I error.

2. **Type II Error**:
   - Type II error occurs when the null hypothesis (H0) is incorrectly not rejected when it is actually false. In other words, it is a false negative.
   - The probability of making a Type II error is denoted by beta (β). Power (1 - β) of the test is the probability of correctly rejecting the null hypothesis when it is false.
   - Example scenario: Continuing with the medical example, let's say the new drug actually has a beneficial effect on patients, but the researcher fails to detect this effect in the study and does not reject the null hypothesis. This would be a Type II error because the researcher incorrectly concludes that the drug is not effective when it actually is.

Q4: Explain Bayes's theorem with an example.

Bayes's theorem is a fundamental concept in probability theory that describes how to update the probability of a hypothesis based on new evidence. It provides a way to revise beliefs or probabilities about an event given prior knowledge and new information. 

The theorem is mathematically represented as:

\[ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} \]

Where:
- \( P(A|B) \) is the probability of event A given that event B has occurred (the posterior probability).
- \( P(B|A) \) is the probability of event B given that event A has occurred (the likelihood).
- \( P(A) \) is the prior probability of event A.
- \( P(B) \) is the prior probability of event B.

Let's illustrate Bayes's theorem with an example:

Suppose a factory produces widgets, and the quality of these widgets can be either "good" or "defective". Historical data suggests that 90% of the widgets produced are good, while 10% are defective.

Now, a quality control inspector selects a widget randomly and tests it. The test correctly identifies a defective widget 95% of the time (true positive rate), but it also wrongly classifies a good widget as defective 3% of the time (false positive rate).

Let's denote:
- A: The widget is good.
- B: The widget is classified as defective.

We want to find the probability that a randomly selected widget is actually good (A), given that it's classified as defective (B).

Using Bayes's theorem:
\[ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} \]

Here,
- \( P(A) = 0.9 \) (prior probability that a widget is good)
- \( P(B|A) = 0.03 \) (false positive rate, probability of a good widget being classified as defective)
- \( P(B|\neg A) = 0.95 \) (true positive rate, probability of a defective widget being classified as defective)
- \( P(B) = P(B|A) \times P(A) + P(B|\neg A) \times P(\neg A) \)
   \( = 0.03 \times 0.9 + 0.95 \times 0.1 \)
   \( = 0.027 + 0.095 \)
   \( = 0.122 \)

Now, applying Bayes's theorem:
\[ P(A|B) = \frac{0.03 \times 0.9}{0.122} \]
\[ P(A|B) = \frac{0.027}{0.122} \]
\[ P(A|B) ≈ 0.221 \]

So, given that the widget is classified as defective, there's approximately a 22.1% chance that it's actually good. This illustrates how Bayes's theorem helps update our beliefs about the quality of the widgets based on the results of the test.

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

A confidence interval is a range of values that is likely to contain the true population parameter, such as a population mean or proportion, with a certain level of confidence. It quantifies the uncertainty associated with estimating a population parameter from a sample.

Here's how to calculate a confidence interval:

1. **Determine the Sample Statistic**: Calculate the sample statistic (e.g., sample mean, sample proportion) from the data.

2. **Select the Confidence Level**: Choose a confidence level, typically denoted by \(1 - \alpha\), where \(\alpha\) is the significance level (often set to 0.05 or 0.01). Common confidence levels include 90%, 95%, and 99%.

3. **Find the Critical Value(s)**: Determine the critical value(s) from the appropriate probability distribution (e.g., z-table for a z-distribution, t-table for a t-distribution) corresponding to the chosen confidence level and the degrees of freedom (if using a t-distribution).

4. **Calculate the Margin of Error**: Compute the margin of error using the formula:
   \[ \text{Margin of Error} = \text{Critical Value} \times \text{Standard Error} \]

5. **Construct the Confidence Interval**: Construct the confidence interval by adding and subtracting the margin of error from the sample statistic:
   \[ \text{Confidence Interval} = \text{Sample Statistic} \pm \text{Margin of Error} \]

Now, let's illustrate with an example:

Suppose we want to estimate the average height of students in a school. We take a random sample of 50 students and measure their heights. The sample mean height is 65 inches, and the sample standard deviation is 3 inches.

Let's calculate a 95% confidence interval for the population mean height.

1. Sample Statistic: Sample mean (\(\bar{x}\)) = 65 inches

2. Confidence Level: \(1 - \alpha = 0.95\) (95% confidence level)

3. Critical Value: Since the sample size is large (n = 50), we can use the z-distribution. For a 95% confidence level, the critical z-value is approximately 1.96 (obtained from a standard normal distribution table).

4. Margin of Error: 
   \[ \text{Margin of Error} = 1.96 \times \frac{3}{\sqrt{50}} \] (using the sample standard deviation to estimate the population standard deviation)

5. Calculate Margin of Error:
   \[ \text{Margin of Error} ≈ 1.96 \times \frac{3}{\sqrt{50}} ≈ 0.83 \]

6. Construct Confidence Interval:
   \[ \text{Confidence Interval} = 65 \pm 0.83 \]
   \[ \text{Confidence Interval} ≈ (64.17, 65.83) \]

Therefore, we can be 95% confident that the true population mean height of students in the school falls within the range of 64.17 to 65.83 inches.

- A: The event that the person has the disease.
- B: The event that the test result is positive.

We want to find \( P(A|B) \), the probability that the person has the disease given a positive test result.

Using Bayes' Theorem:
\[ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} \]

Given:
- \( P(A) = 0.001 \) (probability of having the disease)
- \( P(B|A) = 0.99 \) (probability of a positive test result given that the person has the disease)
- \( P(\neg A) = 1 - P(A) = 0.999 \) (probability of not having the disease)
- \( P(B|\neg A) = 0.05 \) (probability of a positive test result given that the person does not have the disease)

First, let's calculate \( P(B) \), the total probability of getting a positive test result:
\[ P(B) = P(B|A) \times P(A) + P(B|\neg A) \times P(\neg A) \]
\[ P(B) = 0.99 \times 0.001 + 0.05 \times 0.999 \]
\[ P(B) = 0.00099 + 0.04995 \]
\[ P(B) = 0.05094 \]

Now, applying Bayes' Theorem:
\[ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} \]
\[ P(A|B) = \frac{0.99 \times 0.001}{0.05094} \]
\[ P(A|B) ≈ \frac{0.00099}{0.05094} \]
\[ P(A|B) ≈ 0.0194 \]

So, given that the test result is positive, the probability that the person actually has the disease is approximately 0.0194, or about 1.94%. This illustrates how Bayes' Theorem can be used to update the probability of an event based on new evidence.

Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

The margin of error in a confidence interval represents the amount of uncertainty or variability associated with estimating a population parameter (such as a mean or proportion) based on a sample statistic. It indicates the range within which the true population parameter is likely to lie.

The margin of error is calculated using the standard error of the sample statistic and is influenced by two main factors: the standard deviation of the population and the sample size. A larger sample size typically results in a smaller margin of error.

Here's how sample size affects the margin of error:
- As the sample size increases, the standard error decreases. This is because larger sample sizes provide more information about the population, leading to a more precise estimate of the population parameter.
- A smaller standard error means a smaller margin of error, as the estimate becomes more accurate.

Example scenario:

- The confidence interval for the average height might be wider, indicating more uncertainty about the true population mean.

For the larger sample size:
- With a larger sample size, there is less variability in the sample mean, resulting in a smaller margin of error.
- The confidence interval for the average height would be narrower, indicating greater precision in estimating the true population mean.

In summary, larger sample sizes lead to smaller margins of error because they provide more reliable estimates of the population parameter, reducing uncertainty and increasing precision in statistical inference.

Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.

The z-score (also known as the standard score) is a measure of how many standard deviations a data point is away from the mean of the population. It is calculated using the formula:

\[ z = \frac{x - \mu}{\sigma} \]

Where:
- \( x \) is the value of the data point.
- \( \mu \) is the population mean.
- \( \sigma \) is the population standard deviation.

Given:
- \( x = 75 \) (value of the data point)
- \( \mu = 70 \) (population mean)
- \( \sigma = 5 \) (population standard deviation)

Substituting the values into the formula:

\[ z = \frac{75 - 70}{5} \]
\[ z = \frac{5}{5} \]
\[ z = 1 \]

Interpretation:
The calculated z-score of 1 indicates that the data point with a value of 75 is 1 standard deviation above the mean of the population. In other words, it is located one standard deviation unit away from the mean in the positive direction. This means that the data point is relatively higher than the average value in the population by one standard deviation.

Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

To conduct a hypothesis test to determine if the weight loss drug is significantly effective, we'll use a t-test since we're dealing with a small sample size (n = 50) and the population standard deviation is unknown.

Here are the steps for conducting the hypothesis test:

1. **State the Hypotheses**:
   - Null Hypothesis (H0): The average weight loss with the drug is not significantly different from zero (no effect).
   - Alternative Hypothesis (H1): The average weight loss with the drug is significantly different from zero (there is an effect).

   Mathematically:
   - H0: μ = 0 (no effect)
   - H1: μ ≠ 0 (there is an effect)

2. **Choose the Significance Level**:
   - A 95% confidence level corresponds to a significance level (α) of 0.05.

3. **Calculate the Test Statistic**:
   - We'll use the t-statistic formula:
     \[ t = \frac{\bar{x} - \mu_0}{\frac{s}{\sqrt{n}}} \]
     where:
     - \( \bar{x} \) is the sample mean (average weight loss)
     - \( \mu_0 \) is the hypothesized population mean under the null hypothesis (in this case, 0)
     - \( s \) is the sample standard deviation
     - \( n \) is the sample size

4. **Determine the Critical Value**:
   - Degrees of freedom (df) = \( n - 1 \) = 50 - 1 = 49
   - Find the critical t-value for a two-tailed test with 49 degrees of freedom and a significance level of 0.05.

5. **Compare the Test Statistic with the Critical Value**:
   - If the absolute value of the test statistic is greater than the critical value, we reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.

Let's perform the calculations:

\[ t = \frac{6 - 0}{\frac{2.5}{\sqrt{50}}} \]
\[ t = \frac{6}{\frac{2.5}{\sqrt{50}}} \]
\[ t = \frac{6}{\frac{2.5}{7.07}} \]
\[ t ≈ \frac{6}{0.889} \]
\[ t ≈ 6.75 \]

For a two-tailed test at a 95% confidence level with 49 degrees of freedom, the critical t-value is approximately ±2.009 (obtained from a t-distribution table).

Since the absolute value of the calculated t-statistic (6.75) is greater than the critical value (2.009), we reject the null hypothesis.

Conclusion:
Based on the results of the t-test, we have sufficient evidence to conclude that the weight loss drug is significantly effective at a 95% confidence level.
