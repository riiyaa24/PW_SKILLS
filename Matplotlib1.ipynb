Q1. What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used for creating high-quality plots and charts in various formats such as line plots, scatter plots, bar plots, histograms, pie charts, and more.


1. **Flexibility**: Matplotlib provides extensive customization options for plots and charts, allowing users to create a wide range of visualizations according to their needs.

2. **Publication-quality output**: Matplotlib produces high-quality graphics suitable for publication in scientific journals, reports, and presentations.

3. **Wide range of plots**: Matplotlib supports a variety of plot types, making it suitable for visualizing different types of data and relationships.

4. **Integration with other libraries**: Matplotlib can be easily integrated with other Python libraries such as NumPy, Pandas, and SciPy, allowing users to create sophisticated visualizations from their data.

5. **Active development and community support**: Matplotlib is actively developed and maintained, with a large and supportive community providing resources, tutorials, and documentation.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

1. **Line plot**: Used to visualize the relationship between two variables by plotting data points and connecting them with lines.

2. **Scatter plot**: Used to display the relationship between two variables by plotting individual data points on a Cartesian plane.

3. **Bar plot**: Used to compare categorical data by representing each category as a bar, with the height of the bar corresponding to the value of the category.

4. **Histogram**: Used to visualize the distribution of a continuous variable by dividing the data into bins and plotting the frequency of observations in each bin.

5. **Pie chart**: Used to represent categorical data as a circular chart divided into slices, with each slice representing a category and the size of the slice corresponding to the proportion of the category relative to the total.

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Plotting the lines
plt.plot(x1, y1, label='Line 1')
plt.plot(x2, y2, label='Line 2')
plt.plot(x3, y3, label='Line 3')
plt.plot(x4, y4, label='Line 4')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines Plot')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Plotting a bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Profit of Companies')
plt.show()

# Plotting a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightgreen')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Profit of Companies')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Box 1 and Box 2')
plt.grid(True)
plt.show()
```
