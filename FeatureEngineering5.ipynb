Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

To calculate the Pearson correlation coefficient between the amount of time students spend studying for an exam and their final exam scores, you would typically follow these steps:

1. Collect Data: Gather data on the amount of time spent studying (in hours) and the corresponding final exam scores for a sample of students.

2. Calculate Means: Compute the mean (average) of the study time and final exam scores.

3. Calculate Covariance: Calculate the covariance between study time and exam scores.

4. Calculate Standard Deviations: Compute the standard deviations of study time and exam scores.

5. Compute Pearson Correlation Coefficient: Use the formula for the Pearson correlation coefficient:

\[
\rho = \frac{{\text{{cov}}(X, Y)}}{{\sigma_X \cdot \sigma_Y}}
\]

Where:
- \(\rho\) is the Pearson correlation coefficient.
- \(\text{{cov}}(X, Y)\) is the covariance between study time (X) and exam scores (Y).
- \(\sigma_X\) is the standard deviation of study time.
- \(\sigma_Y\) is the standard deviation of exam scores.

6. Interpret the Result: The Pearson correlation coefficient (\(\rho\)) ranges from -1 to 1. Here's how to interpret the result:
   - If \(\rho\) is close to 1, it indicates a strong positive linear relationship between study time and exam scores. This means that as study time increases, exam scores tend to increase.
   - If \(\rho\) is close to -1, it indicates a strong negative linear relationship between study time and exam scores. This means that as study time increases, exam scores tend to decrease.
   - If \(\rho\) is close to 0, it indicates no linear relationship between study time and exam scores.

Suppose you have collected data on the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two variables and interpret the result.

To calculate Spearman's rank correlation coefficient between the amount of sleep individuals get each night and their overall job satisfaction level, you would typically follow these steps:

3. Calculate the Difference in Ranks: For each pair of observations, calculate the difference in ranks between the two variables.

\[

3. **Calculate Covariance**: Calculate the covariance (\(cov(X, Y)\)) between the number of sales calls made per day and the number of sales made per week using the formula:

\[
cov(X, Y) = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{n - 1}
\]

where:
- \(X_i\) and \(Y_i\) are the individual values of the variables.

- \(n\) is the number of observations (in this case, 30).

4. **Calculate Standard Deviations**: Compute the standard deviations (\(\sigma_X\) and \(\sigma_Y\)) of the number of sales calls made per day and the number of sales made per week.

5. **Compute Pearson Correlation Coefficient**: Use the formula for the Pearson correlation coefficient (\(\rho\)):

\[
\rho = \frac{cov(X, Y)}{\sigma_X \cdot \sigma_Y}
\]
